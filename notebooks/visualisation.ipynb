# `oncoj.visual` вҖ” Visualisation Notebook

Explores the `src/oncoj/visual` package, starting with the ASCII syntax-tree renderer.

**Current visualisation API**

| Function | Signature | Description |
|---|---|---|
| `ascii_tree` | `(utt, *, show_comments=True, show_annotations=True, colour=False) вҶ’ str` | Render one `Utterance` as an ASCII tree string |
| `print_tree` | `(utt, *, show_comments=True, show_annotations=True, colour=None)` | Print `ascii_tree(utt)` to stdout |

**Tree layout**
- Title line: sentence ID (bold when colour on) and header word list
- Comment lines: original-script glosses, prefixed with `#` (suppressed by `show_comments=False`)
- Every node's syntactic tag is directly attached to the graph edge; sibling indices (`;@N`) are not shown
- All annotated nodes use the same format: `TAG  ( annotationsвҖҰ )` (suppressed by `show_annotations=False`)
  - **Leaf nodes**: `TAG  ( form  phon_tag  [lemma] )`
  - **Internal nodes with lemma** (compounds, MK groupings): `TAG  ( [lemma] )`

**Colour scheme** (when `colour=True`)

| Part | Style |
|---|---|
| Syntactic tag | **bold** |
| Word form | cyan |
| Phon/script tag (`LOG`, `PHON`, вҖҰ) | dim |
| Lemma ID `[LвҖҰ]` | yellow |

## 0. Setup

```python
import sys
import os

sys.path.insert(0, os.path.join('..', 'src'))

from oncoj.core.corpus import CorpusDocument
from oncoj.visual import ascii_tree, print_tree

TREES_DIR = os.path.join('..', 'data', 'xml', 'trees')
TEXT_DIR  = os.path.join('..', 'data', 'xml', 'text')

print('Python:', sys.version.split()[0])
```

## 1. Basic Usage

Load a corpus file and print a single utterance.

```python
doc = CorpusDocument.from_file(os.path.join(TREES_DIR, 'MYS_01.xml'))
print(f'Loaded {doc.filename}: {len(doc)} utterances')

# MYS.1.4 вҖ” a short poem with a manageable tree
utt = doc[3]
print_tree(utt)
```

---
## 2. Rendering Options

`ascii_tree` and `print_tree` share three keyword arguments:

| Argument | Default | Effect |
|---|---|---|
| `show_comments` | `True` | Include `# вҖҰ` comment lines (original-script glosses) |
| `show_annotations` | `True` | Show word-level annotations `( form  phon  [lemma] )` on nodes |
| `colour` (`ascii_tree`) | `False` | Apply ANSI colour to returned string |
| `colour` (`print_tree`) | `None` | Auto-detect TTY; `True`/`False` to override |

### `show_comments=False` вҖ” hide comment lines

```python
print_tree(utt, show_comments=False)
```

### `show_annotations=False` вҖ” pure structural tree (tags only)

```python
print_tree(utt, show_annotations=False)
```

### `colour=True` вҖ” ANSI highlighting

`print_tree` auto-detects TTY (`colour=None`, the default). Pass `colour=True` to force it on
(useful in notebooks or scripts that capture output for a terminal).
`ascii_tree` defaults to `colour=False` so captured strings stay plain.

```python
print_tree(utt, colour=True, show_comments=False)
```

### `ascii_tree` returns a string

`ascii_tree()` returns a plain string вҖ” useful for slicing, searching, or writing to a file.

```python
tree_str = ascii_tree(utt)

lines = tree_str.splitlines()
print(f'Total lines: {len(lines)}')
print()

# Lines containing a lemma annotation
annotated = [l for l in lines if '[L' in l or '[N' in l]
print(f'Lines with lemma annotations: {len(annotated)}')
for l in annotated:
    print(' ', l.strip())
```

## 3. Browsing by Sentence ID

Look up and render an utterance by its corpus ID.

```python
def get_by_id(doc: CorpusDocument, sentence_id: str):
    for utt in doc:
        if utt.sentence_id == sentence_id:
            return utt
    raise KeyError(sentence_id)

utt_14 = get_by_id(doc, 'MYS.1.10')
print_tree(utt_14)
```

---
## 4. Printing Multiple Trees

Render several utterances from the same document, separated by a rule.

```python
RULE = 'в”Җ' * 60

# Pick short poems (вүӨ 20 corpus lines) from the first 20 utterances
short = [utt for utt in list(doc)[:20] if len(utt.corpus_lines()) <= 20]

for utt in short[:3]:
    print(RULE)
    print_tree(utt)
    print()
```

## 5. Finding Utterances by Word Form or Lemma

Use `CorpusDocument.find_by_form` / `find_by_lemma` to locate matches,
then visualise the containing utterance.

```python
# Find utterances that contain the word 'ama'
hits = doc.find_by_form('ama')
print(f"'ama' appears in {len(hits)} corpus line(s)")

# Collect unique utterances that contain it
seen_ids = set()
unique_utts = []
for utt, cl in hits:
    if utt.sentence_id not in seen_ids:
        seen_ids.add(utt.sentence_id)
        unique_utts.append(utt)

print(f'Unique utterances: {len(unique_utts)}')
print()

# Print the first match
print_tree(unique_utts[0])
```

```python
# Find utterances containing lemma L000520 (genitive particle 'no')
hits_lemma = doc.find_by_lemma('L000520')
print(f'L000520 appears in {len(hits_lemma)} corpus line(s)')

seen_ids2 = set()
unique_utts2 = []
for utt, cl in hits_lemma:
    if utt.sentence_id not in seen_ids2:
        seen_ids2.add(utt.sentence_id)
        unique_utts2.append((utt, len(utt.corpus_lines())))

# Show utterances sorted by size so we can pick a compact one to visualise
unique_utts2.sort(key=lambda x: x[1])
print(f'Unique utterances: {len(unique_utts2)}')
print()

# Visualise the shortest one that has L000520
print_tree(unique_utts2[0][0])
```

## 6. Across Multiple Files

Scan an entire directory and render a tree from each file.

```python
def shortest_utt(doc: CorpusDocument, min_lines: int = 5) -> object:
    """Return the shortest utterance with at least min_lines corpus lines."""
    candidates = [(len(u.corpus_lines()), u) for u in doc
                  if len(u.corpus_lines()) >= min_lines]
    return min(candidates, key=lambda x: x[0])[1] if candidates else None

RULE = 'в”Җ' * 60

for fname in sorted(os.listdir(TREES_DIR))[:5]:
    if not fname.endswith('.xml'):
        continue
    d = CorpusDocument.from_file(os.path.join(TREES_DIR, fname))
    utt = shortest_utt(d)
    if utt is None:
        continue
    print(RULE)
    print(f'File: {fname}  ({len(d)} utterances)')
    print()
    print_tree(utt)
    print()
```

## 7. Field-List Mode (Inline Utterances)

`ascii_tree` also works on `Utterance` objects built from raw field-list lines
(the old comma-path format), not just XML-backed ones.

```python
from oncoj.core.corpus import Utterance, CorpusLine, CommentLine

# Construct a minimal utterance in field-list mode
utt_manual = Utterance([
    CommentLine('=N(" kamu nu ")'),
    CorpusLine.parse('IP-MAT,NP,N,LOG,kamu'),
    CorpusLine.parse('IP-MAT,NP,N,L000006a,LOG,nu'),
    CommentLine('ID,manual_01'),
])

print_tree(utt_manual)
```
